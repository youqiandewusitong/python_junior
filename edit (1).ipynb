```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
```

```python
def get_train_data(file_train_path):
    data = pd.read_csv(file_train_path)
    data_x = []
    data_y = []
    for items in data.values:
        data_x.append([float(i) for i in items[1].split(',')])
        data_y.append(items[2])

    return data_x, data_y

def get_test_data(file_test_path):
    data = pd.read_csv(file_test_path)
    pred_id = []
    pred_x = []
    for items in data.values:
        pred_id.append(items[0])
        pred_x.append([float(i) for i in items[1].split(',')])

    return pred_id, pred_x

def knn_pre(train_file, input_file, output_file):
    pred_id, pred_x = get_test_data(input_file)

    x_train, y_train = get_train_data(train_file)

    knn = KNeighborsClassifier(n_neighbors=1)
    knn.fit(x_train, y_train)
    pred_y = knn.predict(pred_x)

    result_all = []

    for i in range(len(pred_id)):
        result = []
        result.append(pred_id[i])
        if str(pred_y[i]) == '0.0':
            result.append(1)
            result.append(0)
            result.append(0)
            result.append(0)
        elif str(pred_y[i]) == '1.0':
            result.append(0)
            result.append(1)
            result.append(0)
            result.append(0)
        elif str(pred_y[i]) == '2.0':
            result.append(0)
            result.append(0)
            result.append(1)
            result.append(0)
        elif str(pred_y[i]) == '3.0':

            result.append(0)
            result.append(0)
            result.append(0)
            result.append(1)
        else:
            pass
        result_all.append(result)

    submit = pd.DataFrame(result_all)
    submit.columns = ['id', 'label_0', 'label_1', 'label_2', 'label_3']
    submit.to_csv(output_file, index=False)

    return
```
